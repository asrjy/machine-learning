## Creating Datasets and using DataLoader

```python
# Dataset can be used to combine data points and labels, DataLoader can be used to create batches from a dataset tensor
from torch.utils.data import Dataset, DataLoader
```

```python
# we can create a dataloader from a normal tensor as well. when we iterate over it, we get values from it each time. 
# the number of values depends on the batch_size parameter when creating the DataLoader
t = torch.arange(7, dtype = torch.float32)
dataloader = DataLoader(t)
for d in dataloader:
    print(d.numpy())
```

```python
# we can specify the batch_size. If there are more values than perfectly divisible, there's an option to drop those 
# remaining values. if they are not dropped, last batch will have a different number of samples
dataloader = DataLoader(t, batch_size = 2, drop_last = True)
for d in dataloader:
    print(d.numpy())
```

```python
# creating random value tensors with integer values
# here x contains the independant variables and y contains the labels.
x = torch.randint(0, 10, (10, 3), dtype=torch.int)
y = torch.randint(0, 10, (10, 1), dtype=torch.int)
```

```python
# two ways to create a dataset. one is using a class like below. it needs to have these three functions
# __init__ where any logic can be applied on the dataset before creation. 
# __len__ that returns the number of samples in the dataset. basically return either length of x or y. 
# __getitem__ that returns the values of x and y at a specific index. 
class JointDataset(Dataset):
    def __init__(self, x, y):
        # where the intial logic happens, such as reading existing arrays, loading a file, filtering data etc.,
        self.x = x
        self.y = y
    def __len__(self):
        return len(self.x)
    def __getitem__(self, idx):
        # returns corresponding sample to given index
        return self.x[idx], self.y[idx]
```

```python
# the created dataset will return both x and corresponding y values in each iteration. 
joineddata = JointDataset(x, y)
for join in joineddata:
    print(join[0], join[1])
```

```python
# instead of creating the above class, ie., losing any customizability of the dataset, we 
# can use the TensorDataset function instead. we need to pass the x and y tensors to it and it'll take care 
# of the rest. 
from torch.utils.data import TensorDataset

joineddata = TensorDataset(x, y)
```

```python
# now we can pass these datasets to DataLoader and specify batch_size and we get samples in each iteration
# that are good to go for modelling
dataloader = DataLoader(joineddata, batch_size = 3, shuffle = True, drop_last = True)
for i, batch in enumerate(dataloader):
    print(f"{i}th batch - {batch}")
```

```python
# we can reuse dataloader. so each epoch we get two two each time. 
for epoch in range(2):
    print(f'{epoch} epoch:')
    for i, batch in enumerate(dataloader, 1):
        print(f"{i} batch:\n {batch}\t, {batch}\n\n\n")
```

```python
class ImageDataset(Dataset):
    def __init__(self, file_list, labels):
        self.file_list = file_list
        self.labels = labels
    def __getitem__(self, index):
        file = self.file_list[index]
        label = self.labels[index]
        return file, label
    def __len__(self):
        return len(self.labels)

image_dataset = ImageDataset(file_list, labels)
for file, label in image_dataset:
    print(file, label)
```

```python
# now we have the image file and the label in a dataset. now we need to perform any operations
# on this data. like convert them all to the same shape and such. 
# we use torchvision to perform this since it is an image dataset

import torchvision.transforms as transforms
img_h, img_w = 80, 80

transform = transforms.Compose([
    transforms.ToTensor(),
    transforms.Resize((img_h, img_w)),
    
])
```

```python
class ImageDataset(Dataset):
    def __init__(self, file_list, labels, transform = None):
        self.file_list = file_list
        self.transform = transform
        self.labels = labels
    def __getitem__(self, index):
        img = Image.open(self.file_list[index])
        if self.transform is not None:
            img = self.transform(img)
        label = self.labels[index]
        return img, label
    def __len__(self):
        return len(self.labels)

image_dataset = ImageDataset(file_list, labels, transform)
```

### Building NN model in PyTorch

```python
# creating a sample dataset
X_train = np.arange(10, dtype='float32').reshape((10, 1))
y_train = np.array([1.0, 1.3, 3.1, 2.0, 5.0, 6.3, 6.6,7.4, 8.0,9.0], dtype='float32')
plt.plot(X_train, y_train, 'o', markersize = 5)
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

```python
# standardize = mean centering and dividing by standard deviation
# standardizing and converting them to torch datasets

X_train_norm = (X_train - np.mean(X_train))/np.std(X_train)
X_train_norm = torch.from_numpy(X_train_norm)
y_train = torch.from_numpy(y_train)
train_ds = TensorDataset(X_train_norm, y_train)
batch_size = 1
train_dataloader = DataLoader(train_ds, batch_size, shuffle = True)
```

```python
torch.manual_seed(1)
weight = torch.randn(1)
weight.requires_grad_()
bias = torch.zeros(1, requires_grad = True)
def model(xb):
    # returning (wx + b)
    return xb @ weight + bias
```

```python
def loss_fn(input, target):
    # using an mse loss function. mean of squares of error ( difference between input and the predicted value)
    return (input - target).pow(2).mean()
```

```python
# to learn params, we use stochastic gradient descent. 
# torch.optim also has sgd method. but for now, we are manually writing it. 
# sgd needs the gradients to be calculated. rather than calculating them manually, we can use torch.autograd.backward function

learning_rate = .001
num_epochs = 200
log_epochs = 10
for epoch in range(num_epochs):
    for x_batch, y_batch in train_dataloader:
        pred = model(x_batch)
        loss = loss_fn(pred, y_batch)
        loss.backward()
    with torch.no_grad():
        weight -= weight.grad * learning_rate
        bias -= bias.grad * learning_rate
        weight.grad.zero_()
        bias.grad.zero_()
    if epoch % log_epochs == 0:
        print(f"Epoch {epoch} Loss {loss.item(): .4f}")
```

```python
print(f"Final Parameters: {weight.item(), bias.item()}")
```

```python
X_test = np.linspace(0, 9, num = 100, dtype = 'float32').reshape(-1, 1)
X_test_norm = (X_test - X_test.mean())/np.std(X_test)
X_test_norm = torch.from_numpy(X_test_norm)
y_pred = model(X_test_norm)

fig = plt.figure(figsize = (6, 3))
ax = fig.add_subplot(1, 1, 1)
plt.plot(X_train_norm.detach().numpy(), y_train.detach().numpy(), 'o', markersize = 5)
plt.plot(X_test_norm.detach().numpy(), y_pred.detach().numpy(), '--', lw = 3)
plt.legend(['Training Examples', 'Linear Regression'], fontsize = 10)
ax.set_xlabel('x', size = 10)
ax.set_ylabel('y', size = 10)
ax.tick_params(axis = 'both', which = 'major', labelsize = 10)
plt.show()
```

```python
# torch.nn module provides a set of loss functions and torch.optim supports most commonly used optimization algorithms

import torch.nn as nn

learning_rate = .001
loss_fn = nn.MSELoss(reduction = 'mean')
input_size = 1
output_size = 1
model = nn.Linear(input_size , output_size)
optimizer = torch.optim.SGD(model.parameters(), lr = learning_rate)
```

```python
# we can simply call step() method on the optimizer to train the model. 
for epoch in range(num_epochs):
    for x_batch, y_batch in train_dataloader:
        # generate the predictions
        pred = model(x_batch)[:, 0]
        # calculate the loss
        loss = loss_fn(pred, y_batch)
        # compute gradients
        loss.backward()
        # update parameters using gradients
        optimizer.step()
        # reset gradients to zero
        optimizer.zero_grad()
    if epoch % log_epochs == 0:
        print(f"Epoch {epoch} Loss {loss.item(): .4f}")
        
```

```python
print(f"Final Parameters: {model.weight.item(), model.bias.item()}")
```

### MLP to classify flowers from Iris Dataset

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split

iris = load_iris()
X = iris['data']
y = iris['target']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 1./3, random_state = 1)
```

```python
X_train_norm = (X_train - np.mean(X_train))/np.std(X_train)
X_train_norm = torch.from_numpy(X_train_norm).float()
y_train = torch.from_numpy(y_train)
train_ds = TensorDataset(X_train_norm, y_train)
torch.manual_seed(1)
batch_size = 2
train_dl = DataLoader(train_ds, batch_size, shuffle = True)
```

```python
class Model(nn.Module):
    def __init__(self, input_size, hidden_size, output_size):
        super().__init__()
        self.layer1 = nn.Linear(input_size, hidden_size)
        self.layer2 = nn.Linear(hidden_size, output_size)
    def forward(self, x):
        x = self.layer1(x)
        x = nn.Sigmoid()(x)
        x = self.layer2(x)
        # SoftMax since we have multiple classes. 
        x = nn.Softmax(dim = 1)(x)
        return x        
```

```python
input_size = X_train_norm.shape[1]
hidden_size = 16
output_size = 3
model = Model(input_size, hidden_size, output_size)
```

```python
learning_rate = .001
loss_fn = nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(model.parameters(), lr = learning_rate)
```

```python
num_epochs = 100
loss_hist = np.zeros(100)
accuracy_hist = np.zeros(100)
for epoch in range(num_epochs):
    for x_batch, y_batch in train_dl:
        pred = model(x_batch)
        loss = loss_fn(pred, y_batch)
        loss.backward()
        optimizer.step()
        optimizer.zero_grad()
        loss_hist[epoch]  += loss.item() * y_batch.size(0)
        is_correct = (torch.argmax(pred, dim = 1) == y_batch).float()
        accuracy_hist[epoch] += is_correct.mean()
    loss_history[epoch] /= len(train_dl.dataset)
    accuracy_history[epoch] /= len(train_dl.dataset)
```

```python
fig = plt.figure(figsize = (12, 5))

ax = fig.add_subplot(1, 2, 1)
ax.plot(loss_hist, lw = 3)
ax.set_title('Training Loss', size = 10)
ax.set_xlabel('Epoch', size = 10)
ax.tick_params(axis = 'both', which = 'major', labelsize = 10)

ax = fig.add_subplot(1, 2, 2)
ax.plot(accuracy_hist, lw = 3)
ax.set_title('Training Accuracy', size = 10)
ax.set_xlabel('Epoch', size = 10)
ax.tick_params(axis = 'both', which = 'major', labelsize =10)
plt.show()
```

```python
# evaluating the trained model on the test dataset
# we use the same values for mean and std to standardize the data

X_test_norm = X_test - np.mean(X_train) / np.std(X_train)
X_test_norm = torch.from_numpy(X_test_norm).float()
y_test = torch.from_numpy(y_test)

pred_test = model(X_test_norm)
correct = (torch.argmax(pred_test, dim = 1) == y_test).float()
accuaracy = correct.mean()

print(f"Accuracy: {accuaracy}")
```

### Activation Functions

```python
# sigmoid function gives value close to zero if value is highly negative. 
# when output of an activation function is close to zero, nn will learn very slowly. 

# a logistic sigmoid function calculates the probability that a sample x belongs to the postive class. it is only for binary classification. 
# sig(z) = 1/(1 + e^(-z))

import numpy as np

X = np.array([1, 1.4, 2.5])
w = np.array([.4, .3, .5])

def net_input(X, w):
    return np.dot(X, w)

def logistic(z):
    return 1.0/(1 + np.exp(-z))

def logistic_activation(X, w):
    return logistic(net_input(X, w))

print(f'P(y=1|x) = {logistic_activation(X, w):.3f}')
```

```python
 W = np.array([[1.1, 1.2, 0.8, 0.4], [0.2, 0.4, 1.0, 0.2], [0.6, 1.5, 1.2, 0.7]])
# A : data array with shape = (n_hidden_units + 1, n_samples) >>> # note that the first column of this array must be 1
A = np.array([[1, 0.1, 0.4, 0.6]])
Z = np.dot(W, A[0])
y_probas = logistic(Z)
print('Net Input: \n', Z)
print('Output Units:\n', y_probas)
```

```python
# softmax function provides the probability of each class in a multi class setting
# softmax(z) = np.exp(z)/np.sum(np.exp(z))

def softmax(z):
    return np.exp(z)/np.sum(np.exp(z))

y_probs = softmax(Z)
y_probs
```

```python
# comparing with torch's implementation of softmax
torch.softmax(torch.from_numpy(Z), dim = 0)
```

```python
# tanh can be interpreted as rescaled version of softmax function
# advantage of tanh over sigmoid/softmax is it has a broader interval/output spectrum (-1, 1)

np.tanh(Z), torch.tanh(torch.from_numpy(Z))
```

```python
# due to asymptotic behaviour of tanh, ie., it has a range of (-1, 1), there will be no change in the tanh output if the value 
# of z is sufficiently large enough. this means the derivative of activations wrt to net input diminishes as the input becomes large
# for example, the tanh of both values is appx the same and adds very little to the training of the nn

# this is called vanishing gradient problem 

np.tanh(25), np.tanh(20)
```

```python
# relu = max(0, z)
# it's derivative is always 1 for postive numbers. therfore it solves the problem of vanishing gradient problems. 

torch.relu(torch.from_numpy(Z))
```
