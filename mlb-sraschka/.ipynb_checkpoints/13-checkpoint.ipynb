```python
# pytorch uses dynamic computation graphs which are more flexible than static counterparts
# dynamic computation graphs are debugging friendly
# it can work with single or multiple gpu systems and also supports mobile systems. 

# pytorch's computational graphs are directed acyclic graphs (DAGs) 
# a computation graph is simply a network of nodes where each node performs an operation
```

```python
import torch

# using torch.sub() and torch.mul() allows us to pass inputs of various ranks and still get output without having to change anything else
# this function will internally create a compuational graph 
def compute(a, b, c):
    r1 = torch.sub(a, b)
    r2 = torch.mul(r1, 2)
    z = torch.add(r2, c)
    return z
```

```python
print(f"Scalar Input: {compute(torch.tensor(1), torch.tensor(2), torch.tensor(3))}")
print(f"Rank 1 Input: {compute(torch.tensor([1]), torch.tensor([2]), torch.tensor([3]))}")
print(f"Rank 2 Input: {compute(torch.tensor([[1]]), torch.tensor([[2]]), torch.tensor([[3]]))}")
```

```python
# symetry problem in neural network arises when all nodes have the same learned parameters, they don't learn aything
# this can be solved by initializing these tensors randomly
# pytorch allows us to initialize tensors randomly using many different probability distributions

# the intuition behind xavier glorot is to roughly balance the variance of gradients across different layers. 
# otherwise some layers may get too much attention and some layers might get too little attention.

# it follows the following distribution
# uniform(-sqrt(6)/(sqrt(nin + nout)), sqrt(6)/(sqrt(nin + nout)))

# if we want to initialize with a gaussian distribution, the recommended standard deviation is
# (sqrt(2)/(sqrt(nin + nout)))

# pytorch supports both normal and uniform versions of xavier glorot initialization

import torch.nn as nn
torch.manual_seed(1)
w = torch.empty(2, 3)
nn.init.xavier_normal_(w)
w
```

### Computing gradients with automatic differentiation

```python
# gradients can also be used to diagnose why a neural network is giving a specific output to a specific example

# a partial derivative can be understood as rate of change of a function with a set of inputs over the change in a specific input
# a gradient is a vector composed of all inputs' partial derivatives

# pytorch provides context for calculating the gradient at that dependent nodes in the computation graph. 

w = torch.tensor(1., requires_grad = True)
b = torch.tensor(.5, requires_grad = True)
x = torch.tensor([1.4])
y = torch.tensor([2.1])
z = torch.add(torch.mul(w, x), b)
loss = (y-z).pow(2).sum()
loss.backward()
print(f'dL/dw: {w.grad: .3f}')
print(f'dL/db: {b.grad: .3f}')
```

```python
# verifying the gradient value wrt w

# z = wx + b
# loss = (y - z)^2
# => L = (y - wx -b)^2
# => dL/dw = 2 * (y - wx -b) * (-x)
# => dL/db = 2 * (y - wx -b) * (-1)

# pytorch uses reverse accumulation to calculate gradients which is more computationally efficient

2 * (y -z) * (-x), 2 * (y - z) * (-1)
```

### Solving XOR Classification Problem

```python
# we generate a toy dataset of 200 samples, drawn from a uniform distribution between [-1, 1) that has two features x1 and x2
# the value of y = 0 if x1 * x2 <0 else 1

import matplotlib.pyplot as plt
import numpy as np

torch.manual_seed(1)
np.random.seed(1)
x = np.random.uniform(low = -1, high = 1, size = (200, 2))
y = np.ones(len(x))
y[x[:, 0] * x[:, 1] < 0] = 0

n_train = 100
x_train = torch.tensor(x[:n_train, :], dtype = torch.float32)
x_valid = torch.tensor(x[n_train:, :], dtype = torch.float32)
y_train = torch.tensor(y[:n_train], dtype = torch.float32)
y_valid = torch.tensor(y[n_train:], dtype = torch.float32)

fig = plt.figure(figsize = (6, 6))
plt.plot(x[y==0, 0], x[y==0, 1], 'o', alpha = .75, markersize = 10)
plt.plot(x[y==1, 0], x[y==1, 1], '<', alpha = .75, markersize = 10)
plt.xlabel(r'$x_1$', size = 15)
plt.ylabel(r'$x_2$', size = 15)
plt.show()
```

```python
# in practice it's always a good idea to start with a simple model as a baseline

model = nn.Sequential(
    nn.Linear(2, 1),
    nn.Sigmoid()
)
model
```

```python
# cross entropy loss because binary classification
loss_fn = nn.BCELoss()
optimizer = torch.optim.SGD(model.parameters(), lr = .001)
```

```python
from torch.utils.data import TensorDataset, DataLoader

train_ds = TensorDataset(x_train, y_train)
torch.manual_seed(1)
batch_size = 2
train_dl = DataLoader(train_ds, batch_size, shuffle = True)
```

```python
torch.manual_seed(1)
num_epochs = 200

def train(model, num_epochs, train_dl, x_valid, y_valid):
    loss_hist_train = np.zeros(num_epochs)
    accuracy_hist_train = np.zeros(num_epochs)
    loss_hist_valid = np.zeros(num_epochs)
    accuracy_hist_valid = np.zeros(num_epochs)
    
    for epoch in range(num_epochs):
        for x_batch, y_batch in train_dl:
            pred = model(x_batch)[:, 0]
            loss = loss_fn(pred, y_batch)
            loss.backward()
            optimizer.step()
            optimizer.zero_grad()
            loss_hist_train[epoch] += loss.item()
            is_correct = ((pred>=0.5).float() == y_batch).float()
            accuracy_hist_train[epoch] += is_correct.mean()
            
        loss_hist_train[epoch] /= n_train
        accuracy_hist_train[epoch] /= n_train/batch_size
        pred = model(x_valid)[:, 0]
        loss = loss_fn(pred, y_valid)
        loss_hist_valid[epoch] = loss.item()
        is_correct = ((pred>=0.5).float() == y_valid).float()
        accuracy_hist_valid[epoch] += is_correct.mean()
    return loss_hist_train, loss_hist_valid, accuracy_hist_train, accuracy_hist_valid
history = train(model, num_epochs, train_dl, x_valid, y_valid)
```

```python
fig = plt.figure(figsize = (16, 4))
ax = fig.add_subplot(1, 2, 1)
plt.plot(history[0], lw = 4)
plt.plot(history[1], lw = 4)
plt.legend(['Train Loss', 'Validation Loss'], fontsize = 10)
ax.set_xlabel('Epochs', size = 10)
ax.set_ylabel('Loss', size = 10)

ax = fig.add_subplot(1, 2, 2)
plt.plot(history[2], lw = 4)
plt.plot(history[3], lw = 4)
plt.legend(['Train Accuracy', 'Validation Accuracy'], fontsize = 10)
ax.set_xlabel('Epochs', size = 10)
ax.set_ylabel('Accuracy', size = 10)
```

```python
# the universal approximation theorem states that a single layer ffn with large enough number of nodes can approximate arbitrary continuous functions very well. 
# more depth is easier, fewer params and similar performance to wider NNs.
# however they are more prone to vanishing and exploding gradient problems. 

model = nn.Sequential(
    nn.Linear(2, 4),
    nn.ReLU(),
    nn.Linear(4, 4),
    nn.ReLU(), 
    nn.Linear(4, 1),
    nn.Sigmoid()
)

loss_fn = nn.BCELoss()
optimizer = torch.optim.SGD(model.parameters(), lr = .015)

history = train(model, num_epochs, train_dl, x_valid, y_valid)

fig = plt.figure(figsize = (16, 4))
ax = fig.add_subplot(1, 2, 1)
plt.plot(history[0], lw = 4)
plt.plot(history[1], lw = 4)
plt.legend(['Train Loss', 'Validation Loss'], fontsize = 10)
ax.set_xlabel('Epochs', size = 10)
ax.set_ylabel('Loss', size = 10)

ax = fig.add_subplot(1, 2, 2)
plt.plot(history[2], lw = 4)
plt.plot(history[3], lw = 4)
plt.legend(['Train Accuracy', 'Validation Accuracy'], fontsize = 10)
ax.set_xlabel('Epochs', size = 10)
ax.set_ylabel('Accuracy', size = 10)
```

### Making model building more flexible with nn.Module

```python
class MyModule(nn.Module):
    def __init__(self):
        super().__init__()
        l1 = nn.Linear(2, 4)
        a1 = nn.ReLU()
        l2 = nn.Linear(4, 4)
        a2 = nn.ReLU()
        l3 = nn.Linear(4, 1)
        a3 = nn.Sigmoid()
        l = [l1, a1, l2, a2, l3, a3]
        self.module_list = nn.ModuleList(l)
    def forward(self, x):
        for f in self.module_list:
            x = f(x)
        return x    
    def predict(self, x):
        x = torch.tensor(x, dtype = torch.float32)
        pred = self.forward(x)[:, 0]
        return (pred >= .5).float()
```

```python
model = MyModule()
model
```

```python
loss_fn = nn.BCELoss()
optimizer = torch.optim.SGD(model.parameters(), lr = .015)
history = train(model, num_epochs, train_dl, x_valid, y_valid)
```
