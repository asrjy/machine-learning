```python
# pytorch uses dynamic computation graphs which are more flexible than static counterparts
# dynamic computation graphs are debugging friendly
# it can work with single or multiple gpu systems and also supports mobile systems. 

# pytorch's computational graphs are directed acyclic graphs (DAGs) 
# a computation graph is simply a network of nodes where each node performs an operation
```

```python
import torch

# using torch.sub() and torch.mul() allows us to pass inputs of various ranks and still get output without having to change anything else
# this function will internally create a compuational graph 
def compute(a, b, c):
    r1 = torch.sub(a, b)
    r2 = torch.mul(r1, 2)
    z = torch.add(r2, c)
    return z
```

```python
print(f"Scalar Input: {compute(torch.tensor(1), torch.tensor(2), torch.tensor(3))}")
print(f"Rank 1 Input: {compute(torch.tensor([1]), torch.tensor([2]), torch.tensor([3]))}")
print(f"Rank 2 Input: {compute(torch.tensor([[1]]), torch.tensor([[2]]), torch.tensor([[3]]))}")
```

### Computing gradients with automatic differentiation

```python
# gradients can also be used to diagnose why a neural network is giving a specific output to a specific example

# a partial derivative can be understood as rate of change of a function with a set of inputs over the change in a specific input
# a gradient is a vector composed of all inputs' partial derivatives

# pytorch provides context for calculating the gradient at that dependent nodes in the computation graph. 

w = torch.tensor(1., requires_grad = True)
b = torch.tensor(.5, requires_grad = True)
x = torch.tensor([1.4])
y = torch.tensor([2.1])
z = torch.add(torch.mul(w, x), b)
loss = (y-z).pow(2).sum()
loss.backward()
print(f'dL/dw: {w.grad: .3f}')
print(f'dL/db: {b.grad: .3f}')
```

```python
# verifying the gradient value wrt w

# z = wx + b
# loss = (y - z)^2
# => L = (y - wx -b)^2
# => dL/dw = 2 * (y - wx -b) * (-x)
# => dL/db = 2 * (y - wx -b) * (-1)

# pytorch uses reverse accumulation to calculate gradients which is more computationally efficient

2 * (y -z) * (-x), 2 * (y - z) * (-1)
```

### Solving XOR Classification Problem

```python
# we generate a toy dataset of 200 samples, drawn from a uniform distribution between [-1, 1) that has two features x1 and x2
# the value of y = 0 if x1 * x2 <0 else 1

import matplotlib.pyplot as plt
import numpy as np

torch.manual_seed(1)
np.random.seed(1)
x = np.random.uniform(low = -1, high = 1, size = (200, 2))
y = np.ones(len(x))
y[x[:, 0] * x[:, 1] < 0] = 0

n_train = 100
x_train = torch.tensor(x[:n_train, :], dtype = torch.float32)
x_valid = torch.tensor(x[n_train:, :], dtype = torch.float32)
y_train = torch.tensor(y[:n_train], dtype = torch.float32)
y_valid = torch.tensor(y[n_train:], dtype = torch.float32)

fig = plt.figure(figsize = (6, 6))
plt.plot(x[y==0, 0], x[y==0, 1], 'o', alpha = .75, markersize = 10)
plt.plot(x[y==1, 0], x[y==1, 1], '<', alpha = .75, markersize = 10)
plt.xlabel(r'$x_1$', size = 15)
plt.ylabel(r'$x_2$', size = 15)
plt.show()
```

```python
# in practice it's always a good idea to start with a simple model as a baseline

model = nn.Sequential(
    nn.Linear(2, 1),
    nn.Sigmoid()
)
model
```

```python
# cross entropy loss because binary classification
loss_fn = nn.BCELoss()
optimizer = torch.optim.SGD(model.parameters(), lr = .001)
```

```python
from torch.utils.data import TensorDataset, DataLoader

train_ds = TensorDataset(x_train, y_train)
torch.manual_seed(1)
batch_size = 2
train_dl = DataLoader(train_ds, batch_size, shuffle = True)
```

```python
torch.manual_seed(1)
num_epochs = 200

def train(model, num_epochs, train_dl, x_valid, y_valid):
    loss_hist_train = np.zeros(num_epochs)
    accuracy_hist_train = np.zeros(num_epochs)
    loss_hist_valid = np.zeros(num_epochs)
    accuracy_hist_valid = np.zeros(num_epochs)
    
    for epoch in range(num_epochs):
        for x_batch, y_batch in train_dl:
            pred = model(x_batch)[:, 0]
            loss = loss_fn(pred, y_batch)
            loss.backward()
            optimizer.step()
            optimizer.zero_grad()
            loss_hist_train[epoch] += loss.item()
            is_correct = ((pred>=0.5).float() == y_batch).float()
            accuracy_hist_train[epoch] += is_correct.mean()
            
        loss_hist_train[epoch] /= n_train
        accuracy_hist_train[epoch] /= n_train/batch_size
        pred = model(x_valid)[:, 0]
        loss = loss_fn(pred, y_valid)
        loss_hist_valid[epoch] = loss.item()
        is_correct = ((pred>=0.5).float() == y_valid).float()
        accuracy_hist_valid[epoch] += is_correct.mean()
    return loss_hist_train, loss_hist_valid, accuracy_hist_train, accuracy_hist_valid
history = train(model, num_epochs, train_dl, x_valid, y_valid)
```

```python
fig = plt.figure(figsize = (16, 4))
ax = fig.add_subplot(1, 2, 1)
plt.plot(history[0], lw = 4)
plt.plot(history[1], lw = 4)
plt.legend(['Train Loss', 'Validation Loss'], fontsize = 10)
ax.set_xlabel('Epochs', size = 10)
ax.set_ylabel('Loss', size = 10)

ax = fig.add_subplot(1, 2, 2)
plt.plot(history[2], lw = 4)
plt.plot(history[3], lw = 4)
plt.legend(['Train Accuracy', 'Validation Accuracy'], fontsize = 10)
ax.set_xlabel('Epochs', size = 10)
ax.set_ylabel('Accuracy', size = 10)
```

### Making model building more flexible with nn.Module

```python
class MyModule(nn.Module):
    def __init__(self):
        super().__init__()
        l1 = nn.Linear(2, 4)
        a1 = nn.ReLU()
        l2 = nn.Linear(4, 4)
        a2 = nn.ReLU()
        l3 = nn.Linear(4, 1)
        a3 = nn.Sigmoid()
        l = [l1, a1, l2, a2, l3, a3]
        self.module_list = nn.ModuleList(l)
    def forward(self, x):
        for f in self.module_list:
            x = f(x)
        return x    
    def predict(self, x):
        x = torch.tensor(x, dtype = torch.float32)
        pred = self.forward(x)[:, 0]
        return (pred >= .5).float()
```

### Writing custom layers in PyTorch

```python
# assume we are trying to implement a linear layer that computes w(x+e) + b where e refers to a random noise variable. 
# we can also have it behave in different ways depending on whether it's being used during training or testing
# creation is very similar to how you create a nn model
# it has an init constructor and a forward function where the actual functionality of the layer is done

class NoisyLinear(nn.Module):
    def __init__(self, input_size, output_size, noise_stddev = .1):
        super().__init__()
        w = torch.Tensor(input_size, output_size)
        self.w = nn.Parameter(w)
        nn.init.xavier_normal_(self.w)
        b = torch.Tensor(output_size).fill_(0)
        self.b = nn.Parameter(b)
        self.noise_stddev = noise_stddev
    
    def forward(self, x, training = False):
        if training: 
            noise = torch.normal(0.0, self.noise_stddev, x.shape)
            x_new = torch.add(noise, x)
        else:
            x_new = x
        return torch.add(torch.mm(x_new, self.w), self.b)
```

```python
torch.manual_seed(1)
noisy_layer = NoisyLinear(4, 2)
x = torch.zeros(1, 4) 
print(noisy_layer(x, training = True))
print(noisy_layer(x, training = True))
print(noisy_layer(x, training = False))
```

```python
class MyModuleNoisy(nn.Module):
    def __init__(self):
        super().__init__()
        self.l1 = NoisyLinear(2, 4, .07)
        self.a1 = nn.ReLU()
        self.l2 = nn.Linear(4, 4)
        self.a2 = nn.ReLU()
        self.l3 = nn.Linear(4, 1)
        self.a3 = nn.Sigmoid()
    def forward(self, x, training = False):
        x = self.l1(x, training)
        x = self.a1(x)
        x = self.l2(x)
        x = self.a2(x)
        x = self.l3(x)
        x = self.a3(x)
        return x
    def predict(self, x_val):
        x = torch.tensor(x_val, dtype= torch.float32)
        pred = self.forward(x)[:, 0]
        return (pred >= .5).float()
```

```python
torch.manual_seed(1)
model = MyModuleNoisy()
model
```

```python
loss_fn = nn.BCELoss()
optimizer = torch.optim.SGD(model.parameters(), lr = .015)
```

```python
torch.manual_seed(1)
num_epochs = 200
loss_hist_train = np.zeros(num_epochs)
loss_hist_valid = np.zeros(num_epochs)
accuracy_train = np.zeros(num_epochs)
accuracy_valid = np.zeros(num_epochs)

for epoch in range(num_epochs):
    for x, y in train_dl:
        pred = model(x)[:, 0]
        loss = loss_fn(pred, y)
        loss.backward()
        optimizer.step()
        optimizer.zero_grad()
        loss_hist_train[epoch] += loss.item()
        is_true = ((pred >= .5).float() == y).float()
        accuracy_train[epoch] += is_true.mean()
    loss_hist_train /= 100/batch_size
    accuracy_train /= 100/batch_size
    pred = model(x_valid)[:, 0]
    loss = loss_fn(pred, y_valid)
    loss_hist_valid[epoch] = loss.item()
    is_true = ((pred >= .5).float() == y_valid).float()
    accuracy_valid[epoch] = is_true.mean()   
```

```python
fig = plt.figure(figsize = (16, 4))

ax = fig.add_subplot(1, 3, 1)
plt.plot(loss_hist_train, lw = 4)
plt.plot(loss_hist_valid, lw = 4)
ax.set_xlabel('Epoch', size = 10)
ax.set_ylabel('Loss', size = 10)
plt.legend(['Loss - Train', 'Loss - Validation'], fontsize = 10)

ax = fig.add_subplot(1, 3, 2)
plt.plot(accuracy_train, lw = 4)
plt.plot(accuracy_valid, lw = 4)
ax.set_xlabel('Epoch', size = 10)
ax.set_ylabel('Accuracy', size = 10)
plt.legend(['Accuracy - Train', 'Accuracy - Validation'], fontsize = 10)

ax = fig.add_subplot(1, 3, 3)
plot_decision_regions(X = x_valid.numpy(), y = y_valid.numpy().astype(np.int64), clf = model)
ax.set_xlabel(r'$x_1$', size=15)
ax.xaxis.set_label_coords(1, -0.025)
ax.set_ylabel(r'$x_2$', size=15)
ax.yaxis.set_label_coords(-0.025, 1)
plt.show()
```

### Higher Level PyTorch APIs: PyTorch Lightning

```python
# makes building nns with pytorch by removing requirement of writing lot
# of boilerplate code. 
# apart from that allows us to to use multi gpu support and low precision training. 

# the major difference is we use LightningModule instead of nn.Module while defining the model. 

import pytorch_lightning as pl
import torch
import torch.nn as nn

from torchmetrics import Accuracy

class MultiLayerPerceptron(pl.LightningModule):
    def __init__(self, image_shape = (1, 28, 28), hidden_units = (32, 16)):
        super().__init__()
        # pytorch lightning attributes
        self.train_acc = Accuracy()
        self.valid_acc = Accuracy()
        self.test_acc = Accuracy()
        
        # mdoel similar to normal pytorch model creation
        input_size = image_shape[0] * image_shape[1] * image_shape[2]
        all_layers = [nn.Flatten()]
        for hidden_unit in hidden_units:
            layer = nn.Linear(input_size, hidden_unit)
            all_layers.append(layer)
            all_layers.append(nn.ReLU())
            input_size = hidden_unit
        all_layers.append(nn.Linear(hidden_units[-1], 10))
        self.model = nn.Sequential(*all_layers)
    
    def forward(self, x):
        x = self.model(x)
        return x
    
    def training_step(self, batch, batch_idx):
        x, y = batch
        logits = self(x)
        loss = nn.functional.cross_entropy(self(x), x)
        preds = torch.argmax(logits, dim = 1)
        self.valid_acc.update(preds, y)
        self.log("valid_loss", loss, prog_bar = True)
        self.log("valid_acc", self.valid_acc.compute(), prog_bar = True)
        return loss
    
    def test_step(self, batch, batch_idx):
        x, y = batch
        logits = self(x)
        loss = nn.functional.cross_entropy(self(x), y)
        preds = torch.argmax(logits, dim = 1)
        self.test_acc.update(preds, y)
        self.log("test_loss", loss, prog_bar = True)
        self.log("test_acc", self.test_acc.compute(), prog_bar = True)
        return loss

    def configure_optimizers(self):
        optimizer = torch.optim.Adam(self.parameters(), lr = .001)
        return optimizer
        
```
